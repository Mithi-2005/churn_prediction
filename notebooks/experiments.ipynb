```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.model_selection import train_test_split
import pickle
import os
import sys
sys.path.append(os.path.abspath(os.getcwd()))
```

```python
###Load the dataset

df = pd.read_csv('../datasets/churn_modelling.csv')
df.head()
```

```python
## Row number , customerId , surname are not important

df=df.drop(columns=['RowNumber','Surname','CustomerId'],axis=1)
df.head()
```

```python
## Encode categorical variables

le_gender=LabelEncoder()
df['Gender']=le_gender.fit_transform(df['Gender'])
df.head()
```

```python
###OHE for Geography

from sklearn.preprocessing import OneHotEncoder
ohe_geo=OneHotEncoder()
geo_encoder=ohe_geo.fit_transform(df[['Geography']])
geo_encoder
```

```python
geo_encoder=geo_encoder.toarray()
```

```python
geo_encoder
```

```python
ohe_geo.get_feature_names_out(['Geography'])
```

```python
geo_encoder=pd.DataFrame(geo_encoder,columns=ohe_geo.get_feature_names_out(['Geography']))
```

```python
df=pd.concat([df.drop(['Geography'],axis=1),geo_encoder],axis=1)
df.head()
```

```python
with open('../artifacts/label_encoder_gender.pkl','wb') as file:
    pickle.dump(le_gender,file)
with open('../artifacts/ohe_geo.pkl','wb') as file:
    pickle.dump(ohe_geo,file)
```
